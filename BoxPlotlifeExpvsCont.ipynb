```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
def plot(file):
    """
    Description:
        Function for box plot of life exp against continent
    Parameters:
        file: path to .csv file
    returns:
        nothing
    """
    df = extractData(file)
    sns.boxplot(x='continent', y='lifeExp', data=df, palette='coolwarm')
    plt.show()
```

```python
def extractData(file):
    """
    Description:
        Function to extract data from .csv file
    Parameters:
        file: file path
    returns:
        required data
    """
    df = pd.read_csv(file)

    
    return df
```

```python
def main():
    #print(extractData('titanic.csv'))
    plot('gapminder-FiveYearData.csv')
```

```python
if __name__=="__main__":
    main()
```
