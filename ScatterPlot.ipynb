```python
import matplotlib.pyplot as plt 
import pandas as pd
import seaborn as sns
```

```python
def plot(file):
    """
    Description:
        Function for scatter plot of day against totalbill
    Parameters:
        file: path to .csv file
    returns:
        nothing
    """
    df = extractData(file)
    sns.scatterplot(x='day', y='total_bill', data=df, s=100, color="blue", marker="o")
    plt.show()
```

```python
def extractData(file):
    """
    Description:
        Function to extract data from .csv file
    Parameters:
        file: file path
    returns:
        required data
    """
    df = pd.read_csv(file)

    
    return df
```

```python
def main():
    #print(extractData('titanic.csv'))
    plot('tips.csv')
```

```python
if __name__=="__main__":
    main()
```
