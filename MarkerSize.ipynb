```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def plot(x,y,sizes):
    """
    Description: 
        Function to plot scatterplot with different marker sizes
    Parameters:
        x: x aisa random values
        y: y axis random values
        sizes : sizes for marker
    Returns: 
        nothing
    """
    plt.figure(figsize=(10,6))
    plt.scatter(x, y, edgecolor='blue', facecolor='none', alpha=0.7, s=sizes)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
```

```python
def main():
    np.random.seed(0)
    x = np.random.rand(100)*100
    y = np.random.rand(100)*100
    sizes = np.random.rand(100) * 1000

    plot(x,y,sizes)
    
```

```python
if __name__=="__main__":
    main()
```
