```python
import plotly.express as px
import numpy as np

# Generate random x and y coordinates
x = np.random.rand(500)
y = np.random.rand(500)

colors = np.random.rand(500)  # Random colors for each point
sizes = 1000 * np.random.rand(500)  # Random sizes for each point

# Create a scatter plot using Plotly
fig = px.scatter(x=x, y=y,color=colors, size=sizes, title="Scatter Plot of 1000 Random Points", labels={'x': 'X Coordinate', 'y': 'Y Coordinate'})
fig.show()
```
