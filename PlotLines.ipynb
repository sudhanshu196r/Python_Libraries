```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
def plot_graph(x,y1,y2):
    """
    Description:
        This function will plot the graph
    Parameters:
        x: values for x axis
        y1: values for y axis
        y2: values for y axis
    Returns:
        nothing
    """
    plt.figure(figsize=(10, 5))  # Set the figure size

    # Plot the first line
    plt.plot(x, y1, marker='o', color='blue', label='y = x^2')

    # Plot the second line
    plt.plot(x, y2, marker='s', color='green', label='y = x')

    # Label the axes and title
    plt.xlabel('X-axis Label')
    plt.ylabel('Y-axis Label')
    plt.title('Plotting Multiple Lines')

    # Show legend
    plt.legend()

    # Add grid for better readability
    plt.grid()

    # Display the plot
    plt.tight_layout()  # Adjust layout to prevent clipping
    plt.show()
```

```python
def main():
    x = [1,2,3,4,5,6]
    y1 = [1,4,9,16,25,36]
    y2 = [2,4,6,8,10,12]
    plot_graph(x,y1,y2)
```

```python
if __name__=="__main__":
    main()
```
