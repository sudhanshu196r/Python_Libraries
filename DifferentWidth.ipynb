```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
```

```python
def plot_graph(x,y1,y2):
    """
    Description:
        This function will generate lines with different style
    Parameters:
        x and different y axis values for different lines
    Returns:
        nothing
    """

    plt.figure(figsize=(10,5))

    plt.plot(x, y1, linestyle='-',linewidth=0.5, marker='o', color='blue', label='y = x^2')

    plt.plot(x, y2, linestyle=':',linewidth=0.4, marker='o', color='red', label='y = x*2')

    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.title('Plotting Multiple Lines with Different Styles')

    # Show legend
    plt.legend()

    # Add grid for better readability
    plt.grid(True)

    # Display the plot
    plt.tight_layout()  # Adjust layout to prevent clipping
    plt.show()
```

```python
def main():
    x = [1,2,3,4,5]
    y1 = [1,4,9,16,25]
    y2 = [2,4,6,8,10]

    plot_graph(x,y1,y2)
```

```python
if __name__=="__main__":
    main()
```
