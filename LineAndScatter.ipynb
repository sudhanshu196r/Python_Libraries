```python
import plotly.graph_objects as go
import numpy as np

# Generate random data
x = np.linspace(0, 10, 100)  # 100 evenly spaced points between 0 and 10
y1 = np.random.rand(100) * 10  # Random values for scatter plot
y2 = np.random.rand(100)*100  # Sine function for line plot

# Create a scatter plot
scatter_plot = go.Scatter(
    x=x,
    y=y1,
    mode='markers',  # 'markers' for scatter plot
    name='Random Scatter',
    marker=dict(color='blue', size=8)
)

# Create a line plot
line_plot = go.Scatter(
    x=x,
    y=y2,
    mode='lines',  # 'lines' for line plot
    name='Random Line',
    line=dict(color='red', width=2)
)

# Combine the plots in a figure
fig = go.Figure()
fig.add_trace(scatter_plot)
fig.add_trace(line_plot)

# Add layout details
fig.update_layout(
    title='Line and Scatter Plot using Plotly',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    legend_title='Plot Type',
    template='plotly'
)

# Show the figure
fig.show()
```
