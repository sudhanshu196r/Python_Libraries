```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def plot(group1_weights,group1_heights,group2_weights,group2_heights,group3_weights,group3_heights):
    """
    Description: 
        Function to plot scatterplot for multiple groups
    Parameters:
        group(n)_weights: weight of group members
        group(n)_heights: height of group members
    Returns: 
        nothing
    """
    plt.figure(figsize=(10,6))
    
    plt.scatter(group1_weights, group1_heights, color='blue', label='Group 1', marker='o', s=100, alpha=0.7)
    plt.scatter(group2_weights, group2_heights, color='green', label='Group 2', marker='s', s=100, alpha=0.7)
    plt.scatter(group3_weights, group3_heights, color='red', label='Group 3', marker='^', s=100, alpha=0.7)

    # Add titles and labels
    plt.title('Comparison of Weights and Heights Across Three Groups')
    plt.xlabel('Weight (kg)')
    plt.ylabel('Height (cm)')

    
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```

```python
def main():
    group1_weights = [60, 62, 67, 70, 72]
    group1_heights = [150, 160, 155, 170, 175]
    group2_weights = [65, 68, 73, 75, 78]
    group2_heights = [155, 165, 160, 180, 185]
    group3_weights = [70, 72, 77, 80, 85]
    group3_heights = [160, 170, 165, 175, 190]


    plot(group1_weights,group1_heights,group2_weights,group2_heights,group3_weights,group3_heights)
```

```python
if __name__=="__main__":
    main()
```
