```python
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
def plot(x,y):
    """
    Description: 
        Function to plot scatterplot with empty circle marker
    Parameters:
        x: x aisa random values
        y: y axis random values
    Returns: 
        nothing
    """
    plt.figure(figsize=(10,6))
    plt.scatter(x, y, edgecolor='blue', facecolor='none', alpha=0.7, s=80)
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
```

```python
def main():
    np.random.seed(0)
    x = np.random.rand(100)
    y = np.random.rand(100)

    plot(x,y)
    
```

```python
if __name__=="__main__":
    main()
```
