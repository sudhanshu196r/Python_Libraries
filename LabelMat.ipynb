```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd
import logger

log = logger.logger_init('LabelMat')
```

```python
def plotGraph(x,y,xaxis_label,yaxis_label,p_title):
    """
    Description:
        This function will create a graph using the parameters provided
    Parameters:
        x: x axis data
        y: y axis data
        xaxis_label: name for x axis
        yaxis_lable: name for y axis
        p_title: title for garph
    Return:
        Nothing
    """

    plt.plot(x,y)
    plt.title(p_title)
    plt.xlabel(xaxis_label)
    plt.ylabel(yaxis_label)

    plt.show()
```

```python
def main():
    x = [1,2,3,4]
    y = [5,6,7,8]

    xaxis_label = 'X-axis'
    yaxis_label = 'Y-axis'

    p_title = 'First Graph'

    log.info(plotGraph(x,y,xaxis_label,yaxis_label,p_title))
```

```python
if __name__=="__main__":
    main()
```
