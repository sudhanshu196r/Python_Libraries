```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
def read_data(file):
    x_values = []
    y_values = []

    with open(file, 'r') as file:
        for line in file:
            # Split each line into x and y components and convert to float
            x, y = map(float, line.split())
            x_values.append(x)
            y_values.append(y)
    
    return x_values, y_values
```

```python
def plot_graph(file):
    x,y = read_data(file)

    plt.plot(x,y)
    plt.title("File Data")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.show()
```

```python
def main():
    plot_graph('data.txt')
```

```python
if __name__=="__main__":
    main()
```
